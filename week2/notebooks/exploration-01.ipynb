```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('../data/raw/train.csv')
```

```python
plt.style.use("dark_background")
```

```python
list(train['Country_Region'].unique())
```

```python
train.head()
```

```python
train_by_date = train.groupby('Date').sum()
```

```python
fig, ax = plt.subplots(figsize = (30, 10))
ax = sns.lineplot(x = train_by_date.index, y = train_by_date['ConfirmedCases'])
ax.xaxis.set_major_locator(plt.MaxNLocator(10)) # reduce number of xticks
```
